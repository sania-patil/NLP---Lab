```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import re

# Download all required resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')  # FIX

# Input text
text = """
Natural Language Processing is a branch of artificial intelligence that helps computers understand,
interpret, and generate human language. It is widely used in applications such as chatbots, language
translation, sentiment analysis, and speech recognition.
"""

# Preprocessing
text = text.lower()
text = re.sub(r'[^a-zA-Z\s]', '', text)

# Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)

# POS Tagging
pos_tags = pos_tag(tokens)

# Output
print("\nWord → POS Tag mapping:\n")
for word, tag in pos_tags:
    print(f"{word:<15} → {tag}")
```
