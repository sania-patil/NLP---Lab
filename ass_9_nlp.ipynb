```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
import random
import os

tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)


corpus = """
Artificial intelligence is transforming the world of technology and science.
Machine learning algorithms learn patterns from large amounts of data.
Deep learning uses neural networks with many layers to solve complex problems.
Natural language processing enables computers to understand human language.
Recurrent neural networks are designed to work with sequential data.
Long short term memory networks solve the vanishing gradient problem in RNNs.
Text generation is a fascinating application of recurrent neural networks.
Language models predict the next word given a sequence of previous words.
Neural networks are inspired by the structure of the human brain.
Data science combines statistics mathematics and computer science together.
Python is the most popular programming language for machine learning tasks.
TensorFlow and PyTorch are the leading frameworks for deep learning research.
Convolutional neural networks are widely used for image recognition tasks.
Transfer learning allows models trained on one task to be applied to another.
Attention mechanisms have revolutionized natural language processing models.
The transformer architecture powers modern language models like GPT and BERT.
Reinforcement learning trains agents to make decisions through trial and error.
Generative adversarial networks can create realistic images from random noise.
Word embeddings represent words as dense vectors in a continuous vector space.
Sequence to sequence models are used for machine translation and summarization.
Artificial intelligence will continue to advance and reshape human civilization.
Scientists use machine learning to discover new patterns in complex datasets.
Deep neural networks require large amounts of labeled training data to perform well.
The field of computer vision has made remarkable progress in recent years.
Autonomous vehicles use deep learning to perceive and navigate the environment.
Speech recognition systems convert spoken language into written text accurately.
Recommendation systems use collaborative filtering to suggest relevant content.
Clustering algorithms group similar data points together without labeled examples.
Dimensionality reduction techniques help visualize high dimensional data clearly.
The future of artificial intelligence holds great promise for humanity overall.
""".strip().lower()

print(f"Corpus length: {len(corpus)} characters")
print(f"Corpus words : {len(corpus.split())} words\n")

tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts([corpus])

total_words = len(tokenizer.word_index) + 1
print(f"Vocabulary size: {total_words} words\n")


input_sequences = []

for line in corpus.split("\n"):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram = token_list[:i+1]
        input_sequences.append(n_gram)

print(f"Total training sequences: {len(input_sequences)}")

max_seq_len = max(len(s) for s in input_sequences)
input_sequences = np.array(
    tf.keras.preprocessing.sequence.pad_sequences(
        input_sequences, maxlen=max_seq_len, padding="pre"
    )
)
print(f"Padded sequence shape: {input_sequences.shape}\n")

X = input_sequences[:, :-1]
y = input_sequences[:, -1]

y = tf.keras.utils.to_categorical(y, num_classes=total_words)

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}\n")


def build_rnn_model():
    model = models.Sequential(name="RNN_Language_Model")

    model.add(layers.Embedding(
        input_dim=total_words,
        output_dim=100,
        input_length=max_seq_len - 1,
        name="embedding"
    ))

    model.add(layers.LSTM(
        150,
        return_sequences=True,
        dropout=0.2,
        recurrent_dropout=0.2,
        name="lstm_1"
    ))

    model.add(layers.LSTM(
        100,
        dropout=0.2,
        recurrent_dropout=0.2,
        name="lstm_2"
    ))

    model.add(layers.Dense(100, activation="relu", name="dense_hidden"))
    model.add(layers.Dropout(0.2))

    model.add(layers.Dense(total_words, activation="softmax", name="output"))

    return model

model = build_rnn_model()
model.summary()

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = callbacks.EarlyStopping(
    monitor="loss", patience=10,
    restore_best_weights=True, verbose=1
)
reduce_lr = callbacks.ReduceLROnPlateau(
    monitor="loss", factor=0.5, patience=5, verbose=1
)

print("\nTraining RNN Language Model...")
history = model.fit(
    X, y,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(history.history["loss"], color="blue", label="Train Loss")
axes[0].set_title("Loss over Epochs")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Categorical Cross-Entropy Loss")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history.history["accuracy"], color="green", label="Train Accuracy")
axes[1].set_title("Accuracy over Epochs")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Accuracy")
axes[1].legend()
axes[1].grid(True)

plt.suptitle("RNN Language Model — Training History", fontsize=13)
plt.tight_layout()
plt.savefig("rnn_training.png", dpi=150)
plt.show()
print("Saved: rnn_training.png")

def generate_text(seed_text, num_words=20, temperature=1.0):
    """
    Generate text from a seed phrase.

    temperature controls randomness:
      - low  (0.3) = conservative, repetitive but coherent
      - mid  (1.0) = balanced
      - high (1.5) = creative but sometimes incoherent
    """
    result = seed_text.lower()

    for _ in range(num_words):
        token_list = tokenizer.texts_to_sequences([result])[0]

        token_list = tf.keras.preprocessing.sequence.pad_sequences(
            [token_list], maxlen=max_seq_len - 1, padding="pre"
        )

        predictions = model.predict(token_list, verbose=0)[0]

        predictions = np.log(predictions + 1e-10) / temperature
        predictions = np.exp(predictions)
        predictions = predictions / predictions.sum()

        predicted_idx = np.random.choice(len(predictions), p=predictions)

        output_word = ""
        for word, idx in tokenizer.word_index.items():
            if idx == predicted_idx:
                output_word = word
                break

        result += " " + output_word

    return result

print("\n" + "="*60)
print("TEXT GENERATION DEMO")
print("="*60)

test_cases = [
    ("artificial intelligence", 15),
    ("deep learning uses",      15),
    ("neural networks are",     15),
    ("machine learning",        15),
    ("the future of",           15),
]

for seed, n_words in test_cases:
    print(f"\nSeed: '{seed}'")
    print("-" * 50)
    for temp, label in [(0.3, "Conservative"), (0.7, "Balanced"), (1.2, "Creative")]:
        generated = generate_text(seed, num_words=n_words, temperature=temp)
        print(f"  [{label:<12}] {generated}")

final_loss = history.history["loss"][-1]
perplexity = np.exp(final_loss)
print(f"\nFinal Training Loss : {final_loss:.4f}")
print(f"Perplexity          : {perplexity:.2f}  (lower is better)")

print("\n" + "="*60)
print("      RNN LANGUAGE MODEL — SUMMARY")
print("="*60)
print(f"  Task              : Text Generation")
print(f"  Corpus            : AI/Technology text ({len(corpus.split())} words)")
print(f"  Vocabulary Size   : {total_words} words")
print(f"  Training Sequences: {len(input_sequences)}")
print(f"  Model             : Embedding → LSTM(150) → LSTM(100) → Dense")
print(f"  Embedding Dim     : 100")
print(f"  Loss              : Categorical Cross-Entropy")
print(f"  Optimizer         : Adam (lr=0.001)")
print(f"  Final Loss        : {final_loss:.4f}")
print(f"  Perplexity        : {perplexity:.2f}")
print("="*60)
```
