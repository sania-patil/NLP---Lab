```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag, RegexpParser

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

paragraph = """
Natural Language Processing is an important field of artificial intelligence.
It helps machines understand human language and extract meaningful information.
Students are working on various NLP projects.
"""
np_list = []

sentences = sent_tokenize(paragraph)

print("\nSentences:")
print(sentences)

for sentence in sentences:
    print("\nProcessing Sentence:", sentence)

    words = word_tokenize(sentence)
    print("Words:", words)

    tagged_words = pos_tag(words)
    print("POS Tags:", tagged_words)

    grammar = "NP: {<DT>?<JJ>*<NN.*>+}"

    parser = RegexpParser(grammar)
    tree = parser.parse(tagged_words)
    print("Parse Tree:", tree)

    for subtree in tree.subtrees():
        if subtree.label() == 'NP':
            np = " ".join(word for word, tag in subtree)
            np_list.append(np)

print("\nExtracted Noun Phrases:")
for np in np_list:
    print(np)
```
